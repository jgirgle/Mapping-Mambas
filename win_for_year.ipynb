```python
nba_db = pd.read_csv("csv_files/raw_file/NBA Finals and MVP Count.csv")
```

```python
nba_db
```

```python
nba_new = nba_db.rename(columns={"NBA Champion":"name"})
nba_year_df = nba_new.rename(columns={"Year":"year"})
```

```python
nba_year_df
```

```python
wins_db = nba_year_df.assign(value = nba_year_df.groupby('name').cumcount()+1)
```

```python
place_holders = pd.DataFrame([{'year': Year, "name": team} for Year in nba_year_df.year.values for team in nba_year_df['name'].unique()])
```

```python
drop_duplicates = pd.concat([wins_db, place_holders]).drop_duplicates(['year', 'name'])
```

```python
nba_year_df["name"].nunique()
```

```python
fixed_dfs = []
```

```python
for team, df in drop_duplicates.groupby("name"): 
    fixed_df = df.sort_values("year").ffill().fillna(0)[['year', 'name', "value"]].assign(lastValue = lambda x:x.value.shift(1)).fillna(0)
    fixed_dfs.append(fixed_df)
```

```python
total_wins_df = pd.concat(fixed_dfs)
```

```python
total_wins_df.query("`name` == 'Los Angeles Lakers'").tail(15)
```

```python
total_wins_df.to_json('total wins by year.json', orient='records')
```
